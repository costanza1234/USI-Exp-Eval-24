```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def read_csv(file):
    with open(file, "r") as f:
        reader = csv.reader(f)
        data = list(reader)
    return data


data = read_csv("sorting_experiment_results.csv")

# we then convert the data into a pandas dataframe
df = pd.DataFrame(data[1:], columns=data[0])
# Convert 'Time(ns)' column to numeric
df['Time(ns)'] = pd.to_numeric(df['Time(ns)'], errors='coerce')
```

```python
# Hypotesis 1 analysis
# Hypothesis 1: Independent variable = sortedness level of the array.
# Dependent variable = time taken to sort the array.
# Counfounding variables = sorting algorithm used, size of the array, data type.


# Boxplot of the time taken to sort the array for each sortedness level
# On the x axis we have the 4 different sortedness levels (Random, Reversed, FirstHalfSorted, SecondHalfSorted)
# and for each sortedness level we have 4 bars, one for each sorting algorithm
# y axis we have the time taken to sort the array in a logarithmic scale


plt.clf()
plt.close('all')
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x="Sortedness", y="Time(ns)", hue="Algorithm", data=df)
plt.yscale("log")
plt.title("Time taken to sort the array for each sortedness level")
plt.savefig("hip1.png")
plt.show()
```

```python
# Same boxplot but without log scale
plt.clf()
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x="Sortedness", y="Time(ns)", hue="Algorithm", data=df)
plt.title("Time taken to sort the array for each sortedness level")
plt.savefig("hip1-nonLog.png")
plt.show()
```

```python
# Hypotesis 2 analysis
# Hypothesis 2: Independent variable = size of the array.
# Dependent variable = time taken to sort the array.
# Counfounding variables = sorting algorithm, data type, sortedness of the array.

# Boxplot of the time taken to sort the array for each sorting algorithm
# x axis we have the 3 different sizes (100, 1000, 10000), and for each size we have 4 bars, one for each sorting algorithm
# y axis we have the time taken to sort the array in a logarithmic scale

# Header is Algorithm,DataType,DataSize,Sortedness,Time(ns)

# Boxplot
plt.clf()
plt.figure(figsize=(10, 6))
sns.boxplot(x="DataSize", y="Time(ns)", hue="Algorithm", data=df)
plt.yscale("log")
plt.title("Time taken to sort the array for each size")
plt.savefig("hip2.png")
plt.show()
```

```python
# Hypotesis 3 analysis
# Hypothesis 3: Independent variable = data type.
# Dependent variable = time taken to sort the array.
# Counfounding variables = sorting algorithm used, size of the array, sortedness of the array.

# Boxplot of the time taken to sort the array for each data type
# On the x axis we have the 3 different data types (Integer, Float, Double, Long)
# and for each data type we have 4 bars, one for each sorting algorithm
# y axis we have the time taken to sort the array in a logarithmic scale

# Boxplot
plt.clf()
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x="DataType", y="Time(ns)", hue="Algorithm", data=df)
plt.yscale("log")
plt.title("Time taken to sort the array for each data type")
plt.savefig("hip3.png")
plt.show()
```
