```python
# we first open the csv file and read it

import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def read_csv(file):
    with open(file, "r") as f:
        reader = csv.reader(f)
        data = list(reader)
    return data


data = read_csv("sorting_experiment_results.csv")

# we then convert the data into a pandas dataframe
df = pd.DataFrame(data[1:], columns=data[0])
```

```python
# Hypotesis 1 analysis
# Hypothesis 1: Independent variable = sorting algorithm used.
# Dependent variable = time taken to sort the array.
# Counfounding variables = size of the array, sortedness level of the array

# Boxplot of the time taken to sort the array for each sorting algorithm
# x axis we have the 3 different sizes (100, 1000, 10000), and for each size we have 4 bars, one for each sorting algorithm
# y axis we have the time taken to sort the array in a logarithmic scale

# Header is Algorithm,DataType,DataSize,Sortedness,Time(ns)

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x="DataSize", y="Time(ns)", hue="Algorithm", data=df)
plt.yscale("log")
plt.title("Time taken to sort the array for each sorting algorithm")
# plt.savefig("hip1.png")
plt.show()
```

```python
# Hypotesis 2 analysis
# Hypothesis 2: Independent variable = sortedness level of the array.
# Dependent variable = time taken to sort the array.
# Counfounding variables = sorting algorithm used, size of the array,
# sortedness level of the array


# Boxplot of the time taken to sort the array for each sortedness level
# On the x axis we have the 4 different sortedness levels (Random, Reversed, FirstHalfSorted, SecondHalfSorted)
# and for each sortedness level we have 4 bars, one for each sorting algorithm
# y axis we have the time taken to sort the array in a logarithmic scale

# Boxplot
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x="Sortedness", y="Time(ns)", hue="Algorithm", data=df)
plt.yscale("log")
plt.title("Time taken to sort the array for each sortedness level")
# plt.savefig("hip2.png")
plt.show()
```
