```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def read_csv(file):
    with open(file, "r") as f:
        reader = csv.reader(f)
        data = list(reader)
    return data

# Header is Algorithm,DataType,DataSize,Sortedness,Time(ns)


data = read_csv("sorting_experiment_results.csv")

# we then convert the data into a pandas dataframe
df = pd.DataFrame(data[1:], columns=data[0])
# Convert 'Time(ns)' column to numeric
df['Time(ns)'] = pd.to_numeric(df['Time(ns)'], errors='coerce')
df['DataSize'] = pd.to_numeric(df['DataSize'], errors='coerce')
```

```python
# Hypotesis 1 analysis
# Hypothesis 1: Independent variable = sortedness level of the array.
# Dependent variable = time taken to sort the array.
# Counfounding variables = sorting algorithm used, size of the array, data type.


# Boxplot of the time taken to sort the array for each sortedness level
# On the x axis we have the 4 different sortedness levels (Random, Reversed, FirstHalfSorted, SecondHalfSorted)
# and for each sortedness level we have 4 bars, one for each sorting algorithm
# y axis we have the time taken to sort the array in a logarithmic scale

# Clear any previous plots
plt.clf()
plt.close('all')

# Set the theme
sns.set_theme(style="whitegrid")

# Create a figure with a specific size
plt.figure(figsize=(12, 8))

# Use a custom palette for better color distinction
custom_palette = sns.color_palette("Set2")

# Create the boxplot
sns.boxplot(x="Sortedness", y="Time(ns)", hue="Algorithm", data=df, palette=custom_palette)

# Set y-axis to logarithmic scale
plt.yscale("log")

# Add title and labels with improved font size
plt.title("Time taken to sort the array for each sortedness level", fontsize=16)
plt.xlabel("Sortedness Level", fontsize=14)
plt.ylabel("Time (ns)", fontsize=14)

# Improve the legend
plt.legend(title="Algorithm", title_fontsize='13', fontsize='11')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid lines for better readability
plt.grid(True, which="both", linestyle='--', linewidth=0.5)

# Save the plot as a PNG file
plt.savefig("hip1.png", bbox_inches='tight')

# Show the plot
plt.show()
```

```python
# Group by Algorithm, Sortedness, and DataType, then describe the Time(ns) column
grouped = df.groupby(['Algorithm', 'Sortedness', 'DataType'])['Time(ns)'].describe()

# Select the required statistics and rename the columns
summary = grouped[['min', '25%', '50%', '75%', 'max']].reset_index()
summary.columns = ['Algorithm', 'SortOrder', 'DataType', 'min', 'FirstQuartile', 'Median', 'ThirdQuartile', 'max']

# Save the summary table to a CSV file
summary.to_csv("summary_table.csv", index=False)

# Display the summary table
print(summary)
```
